# 03_Exploratory_Data_Analysis

**Project:** Taxi Fare Amount Prediction Using NYC Yellow Taxi Big Data

This notebook loads the cleaned DataFrame from `02_Data_Quality_and_Preprocessing.ipynb` and performs a lightweight exploratory data analysis using a sampled Pandas DataFrame for visualisation.


```python
%run ./02_Data_Quality_and_Preprocessing.ipynb

# Sample a small fraction for quick EDA (≈0.1% of rows)
sample_frac = 0.001
pdf = df_clean.sample(withReplacement=False, fraction=sample_frac, seed=42).toPandas()

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')

# 1. Distribution of fare_amount
plt.figure(figsize=(10,6))
sns.histplot(pdf['fare_amount'], bins=50, kde=True)
plt.title('Fare Amount Distribution')
plt.xlabel('Fare Amount (USD)')
plt.ylabel('Count')
plt.show()

# 2. Trip distance vs fare amount
plt.figure(figsize=(10,6))
sns.scatterplot(x='trip_distance', y='fare_amount', data=pdf, alpha=0.5)
plt.title('Fare vs. Trip Distance')
plt.xlabel('Trip Distance (miles)')
plt.ylabel('Fare Amount (USD)')
plt.show()

# 3. Passengers per trip boxplot
plt.figure(figsize=(8,5))
sns.boxplot(x='passenger_count', y='fare_amount', data=pdf)
plt.title('Fare Amount by Passenger Count')
plt.xlabel('Passenger Count')
plt.ylabel('Fare Amount (USD)')
plt.show()

# 4. Hourly pattern (extract hour)
pdf['pickup_hour'] = pd.to_datetime(pdf['pickup_datetime']).dt.hour
plt.figure(figsize=(10,6))
sns.boxplot(x='pickup_hour', y='fare_amount', data=pdf)
plt.title('Fare Amount by Pickup Hour')
plt.xlabel('Pickup Hour (0‑23)')
plt.ylabel('Fare Amount (USD)')
plt.show()
```
