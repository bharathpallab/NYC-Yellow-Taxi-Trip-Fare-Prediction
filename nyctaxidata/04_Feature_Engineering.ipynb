# 04_Feature_Engineering

**Project:** Taxi Fare Amount Prediction Using NYC Yellow Taxi Big Data

This notebook loads the cleaned DataFrame from `02_Data_Quality_and_Preprocessing.ipynb` and creates additional temporal and categorical features required for modeling. It also assembles the feature vector used by Spark ML models.


```python
%run ./02_Data_Quality_and_Preprocessing.ipynb

from pyspark.sql.functions import hour, dayofweek, month, col

# Temporal features
df_feat = df_clean
    .withColumn('pickup_hour', hour(col('pickup_datetime')))
    .withColumn('pickup_dayofweek', dayofweek(col('pickup_datetime')))
    .withColumn('pickup_month', month(col('pickup_datetime')))

# If latitude/longitude columns were present we could add haversine distance – omitted here.

# Assemble feature vector
from pyspark.ml.feature import VectorAssembler
feature_cols = ['passenger_count', 'trip_distance', 'pickup_hour', 'pickup_dayofweek', 'pickup_month']
assembler = VectorAssembler(inputCols=feature_cols, outputCol='features')
df_feat = assembler.transform(df_feat)
df_feat.cache()
print('Feature DataFrame schema:')
df_feat.printSchema()
print(f'Rows after feature engineering: {df_feat.count():,}')
```
