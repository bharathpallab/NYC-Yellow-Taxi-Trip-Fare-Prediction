# 01_Data_Loading_and_Spark_Setup

**Project:** Taxi Fare Amount Prediction Using NYC Yellow Taxi Big Data

This notebook sets up a Spark session with appropriate configuration and loads all six monthly parquet files into a single DataFrame.


```python
# Spark Session & Configuration
from pyspark.sql import SparkSession
from pyspark import SparkConf
import os, glob, time

conf = SparkConf()\
    .setAppName('TaxiFarePrediction')\
    .set('spark.driver.memory', '4g')\
    .set('spark.executor.memory', '4g')\
    .set('spark.executor.cores', '2')\
    .set('spark.sql.shuffle.partitions', '64')\
    .set('spark.sql.execution.arrow.enabled', 'true')  # improves pandas conversion

spark = SparkSession.builder.config(conf=conf).getOrCreate()
spark.sparkContext.setLogLevel('WARN')
seed = 42
print(f'Spark version: {spark.version}')

# Data loading
data_dir = r'C:\Users\shiva\OneDrive\Desktop\mtechbigdata2\nyctaxidata'
parquet_pattern = os.path.join(data_dir, 'yellow_tripdata_2025-*.parquet')
parquet_files = sorted(glob.glob(parquet_pattern))
print(f'Found {len(parquet_files)} parquet files:')
for f in parquet_files: print('  ', os.path.basename(f))

start = time.time()
df = spark.read.parquet(*parquet_files)
load_time = time.time() - start
print(f'Data loaded in {load_time:.2f} seconds')
df.printSchema()
print(f'Row count: {df.count():,}')
df.cache()
```
