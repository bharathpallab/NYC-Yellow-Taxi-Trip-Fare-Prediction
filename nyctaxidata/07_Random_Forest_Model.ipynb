# 07_Random_Forest_Model

**Project:** Taxi Fare Amount Prediction Using NYC Yellow Taxi Big Data

This notebook loads the feature‑engineered DataFrame from `04_Feature_Engineering.ipynb` and trains a Random Forest Regressor using Spark MLlib. The model is persisted for later evaluation.


```python
%run ./04_Feature_Engineering.ipynb

from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml.evaluation import RegressionEvaluator
import os, time

train_df, test_df = df_feat.randomSplit([0.8, 0.2], seed=42)

rf = RandomForestRegressor(featuresCol='features', labelCol='fare_amount', numTrees=100, maxDepth=12)
start = time.time()
rf_model = rf.fit(train_df)
print(f'Random Forest training time: {time.time() - start:.2f}s')

preds = rf_model.transform(test_df)
evaluator = RegressionEvaluator(labelCol='fare_amount', predictionCol='prediction', metricName='rmse')
rmse = evaluator.evaluate(preds)
print(f'Random Forest Test RMSE: {rmse:.4f}')

model_dir = r'C:\Users\shiva\OneDrive\Desktop\mtechbigdata2\nyctaxidata\models'
os.makedirs(model_dir, exist_ok=True)
rf_path = os.path.join(model_dir, 'random_forest')
rf_model.write().overwrite().save(rf_path)
print(f'Saved Random Forest model to {rf_path}')
```
