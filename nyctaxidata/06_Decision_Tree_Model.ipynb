**Project:** Taxi Fare Amount Prediction Using NYC Yellow Taxi Big Data

This notebook loads the feature‑engineered DataFrame from `04_Feature_Engineering.ipynb` and trains a Decision Tree Regressor using Spark MLlib. The model is saved for later evaluation.


```python
%run ./04_Feature_Engineering.ipynb

from pyspark.ml.regression import DecisionTreeRegressor
from pyspark.ml.evaluation import RegressionEvaluator
import os, time

train_df, test_df = df_feat.randomSplit([0.8, 0.2], seed=42)

dt = DecisionTreeRegressor(featuresCol='features', labelCol='fare_amount', maxDepth=10)
start = time.time()
dt_model = dt.fit(train_df)
print(f'Decision Tree training time: {time.time() - start:.2f}s')

preds = dt_model.transform(test_df)
evaluator = RegressionEvaluator(labelCol='fare_amount', predictionCol='prediction', metricName='rmse')
rmse = evaluator.evaluate(preds)
print(f'Decision Tree Test RMSE: {rmse:.4f}')

model_dir = r'C:\Users\shiva\OneDrive\Desktop\mtechbigdata2\nyctaxidata\models'
os.makedirs(model_dir, exist_ok=True)
dt_path = os.path.join(model_dir, 'decision_tree')
dt_model.write().overwrite().save(dt_path)
print(f'Saved Decision Tree model to {dt_path}')
```
