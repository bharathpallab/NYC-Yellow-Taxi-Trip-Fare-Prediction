```python
%run ./04_Feature_Engineering.ipynb

import os
import pandas as pd

# Ensure output directory exists
output_dir = r'C:\Users\shiva\OneDrive\Desktop\mtechbigdata2\nyctaxidata\tableau_exports'
os.makedirs(output_dir, exist_ok=True)

# 1. Basic fare summary by hour
hourly_fare = df_feat.groupBy('pickup_hour')\
    .agg(
        F.avg('fare_amount').alias('avg_fare'),
        F.sum('fare_amount').alias('total_fare'),
        F.count('*').alias('trip_count')
    )
    .orderBy('pickup_hour')

hourly_fare_pd = hourly_fare.toPandas()
hourly_fare_pd.to_csv(os.path.join(output_dir, 'hourly_fare_summary.csv'), index=False)

# 2. Average fare by passenger count
pax_fare = df_feat.groupBy('passenger_count')\
    .agg(F.avg('fare_amount').alias('avg_fare'), F.sum('fare_amount').alias('total_fare'), F.count('*').alias('trip_count'))
    .orderBy('passenger_count')
pax_fare_pd = pax_fare.toPandas()
pax_fare_pd.to_csv(os.path.join(output_dir, 'passenger_fare_summary.csv'), index=False)

# 3. Revenue by month (to show seasonality)
month_fare = df_feat.groupBy('pickup_month')\
    .agg(F.avg('fare_amount').alias('avg_fare'), F.sum('fare_amount').alias('total_fare'), F.count('*').alias('trip_count'))
    .orderBy('pickup_month')
month_fare_pd = month_fare.toPandas()
month_fare_pd.to_csv(os.path.join(output_dir, 'monthly_fare_summary.csv'), index=False)

print('Exported CSV files to', output_dir)

# Business insight examples (printed for quick reference)
print('--- Top 3 hours by total revenue ---')
hourly_fare_pd.sort_values('total_fare', ascending=False).head(3)
print('--- Passenger count with highest average fare ---')
print(pax_fare_pd.loc[pax_fare_pd['avg_fare'].idxmax()])
```
