# 05_Linear_Regression_Model

**Project:** Taxi Fare Amount Prediction Using NYC Yellow Taxi Big Data

This notebook loads the feature‑engineered DataFrame from `04_Feature_Engineering.ipynb` and trains a Linear Regression model using Spark MLlib. The trained model is persisted for later evaluation.


```python
%run ./04_Feature_Engineering.ipynb

from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
import os, time

# Train‑test split (same seed for reproducibility)
train_df, test_df = df_feat.randomSplit([0.8, 0.2], seed=42)

# Model definition
lr = LinearRegression(featuresCol='features', labelCol='fare_amount')

# Train
start = time.time()
lr_model = lr.fit(train_df)
print(f'Linear Regression training time: {time.time() - start:.2f}s')

# Evaluate on test set
preds = lr_model.transform(test_df)
evaluator = RegressionEvaluator(labelCol='fare_amount', predictionCol='prediction', metricName='rmse')
rmse = evaluator.evaluate(preds)
print(f'Linear Regression Test RMSE: {rmse:.4f}')

# Persist model
model_dir = r'C:\Users\shiva\OneDrive\Desktop\mtechbigdata2\nyctaxidata\models'
os.makedirs(model_dir, exist_ok=True)
lr_path = os.path.join(model_dir, 'linear_regression')
lr_model.write().overwrite().save(lr_path)
print(f'Saved Linear Regression model to {lr_path}')
```
