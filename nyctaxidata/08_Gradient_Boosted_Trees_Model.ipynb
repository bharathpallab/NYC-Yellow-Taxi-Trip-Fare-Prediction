# 08_Gradient_Boosted_Trees_Model

**Project:** Taxi Fare Amount Prediction Using NYC Yellow Taxi Big Data

This notebook loads the feature‑engineered DataFrame from `04_Feature_Engineering.ipynb` and trains a Gradient‑Boosted Trees regressor using Spark MLlib. The trained model is saved for later evaluation.


```python
%run ./04_Feature_Engineering.ipynb

from pyspark.ml.regression import GBTRegressor
from pyspark.ml.evaluation import RegressionEvaluator
import os, time

train_df, test_df = df_feat.randomSplit([0.8, 0.2], seed=42)

gbt = GBTRegressor(featuresCol='features', labelCol='fare_amount', maxIter=50, maxDepth=6)
start = time.time()
gbt_model = gbt.fit(train_df)
print(f'Gradient‑Boosted Trees training time: {time.time() - start:.2f}s')

preds = gbt_model.transform(test_df)
evaluator = RegressionEvaluator(labelCol='fare_amount', predictionCol='prediction', metricName='rmse')
rmse = evaluator.evaluate(preds)
print(f'GBT Test RMSE: {rmse:.4f}')

model_dir = r'C:\Users\shiva\OneDrive\Desktop\mtechbigdata2\nyctaxidata\models'
os.makedirs(model_dir, exist_ok=True)
gbt_path = os.path.join(model_dir, 'gradient_boosted_trees')
gbt_model.write().overwrite().save(gbt_path)
print(f'Saved GBT model to {gbt_path}')
```
