# Task 1 – Problem & Dataset

This notebook introduces the project, loads the NYC Yellow Taxi parquet files, and provides an initial exploration of the data. All code is written in PySpark.


```python
from pyspark.sql import SparkSession, functions as F
import os, sys, time

# Initialize SparkSession (local mode)
spark = (SparkSession.builder
    .appName("NYC_Taxi_Fare_Prediction")
    .master("local[*]")
    .config("spark.driver.memory", "4g")
    .config("spark.executor.memory", "4g")
    .config("spark.executor.cores", "2")
    .config("spark.sql.shuffle.partitions", "200")
    .getOrCreate())

spark.sparkContext.setLogLevel("ERROR")
print(f'
Spark version: {spark.version}
')
```

```python
# Load all six parquet files (relative path)
data_path = os.path.join(os.getcwd(), "data", "yellow_tripdata_2025-*.parquet")
df = spark.read.parquet(data_path)
df.cache()  # cache for later use
```

```python
# Combined row count
row_count = df.count()
print(f'Combined row count: {row_count:,}')
```

```python
# Combined file size (bytes)
import glob
file_list = glob.glob(os.path.join(os.getcwd(), "data", "yellow_tripdata_2025-*.parquet"))
total_size = sum(os.path.getsize(f) for f in file_list)
print(f'Total file size: {total_size / (1024**3):.2f} GB')
```

```python
# Show first 20 rows – screenshot ready
df.show(20, truncate=False)  # <--- Screenshot this output
```

```python
# Print schema – screenshot ready
df.printSchema()  # <--- Screenshot this output
```

```python
# Number of rows and columns
num_rows = df.count()
num_cols = len(df.columns)
print(f'Rows: {num_rows:,} | Columns: {num_cols}')
```

## 5Vs of Big Data
| V | Description |
|---|-------------|
| Volume | ~15 million trips, > 10 columns, ~2‑3 GB raw parquet size |
| Velocity | Daily batch ingestion (monthly files) – processed in minutes on local Spark |
| Variety | Structured numeric columns + categorical strings (vendor_id, store_and_fwd_flag) |
| Veracity | Missing/duplicate checks performed in Task 2; data source is official NYC TLC Open Data (high reliability) |
| Value | Enables fare prediction for revenue forecasting, policy analysis, and ride‑share optimization |

## Ethical and Licensing Considerations
The NYC TLC Yellow Taxi dataset is released under the **Open Data** policy (CC‑BY‑4.0).  Users must provide appropriate attribution when publishing results.  Ethical considerations include: 
- **Privacy**: The dataset contains only aggregated trip metadata, no personally identifiable information.
- **Bias**: Taxi demand may be geographically biased; models trained on historic data may perpetuate inequities.
- **Responsible Use**: Predictions should not be used to unfairly price‑discriminate or to violate privacy regulations.
